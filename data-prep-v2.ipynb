# Mining whatsapp chat (Cañitas)

```python
import pandas as pd
import numpy as np
import pandas as pd
import emoji
import re
from stop_words import get_stop_words 
from ftfy import fix_text
import itertools
```

```python
with open('canitas_chat.txt', 'r') as f:
    # Feed the file text into findall(); it returns a list of all the found strings 
    # (written for android export format)
    strings = re.findall('(\d+.\d+.\d+\s+\d+:+\d+)\s.\s(\w+\s\w+\s\w+|\w+\s\w+|\w+):\s(.*)', f.read())
```

```python
# Create list of dictionaries of messages
list_messages = []
for message in strings:
    dict_ = dict({'TimeStamp': message[0], 'Sender_name': message[1], 'Content': message[2]})
    list_messages.append(dict_)
```

```python
clean_list  = []
number_words = []

# Iterate over messages
for message in list_messages: 
    
    # If there is a content key...
    if 'Content' in message.keys():
        # Look in the content key
        clean_msg = fix_text(message['Content'])
        
        # Format multimedia messages
        if '<' in clean_msg and 'omitid' in clean_msg:
            if 'audio' in clean_msg: clean_msg = '<audio>'
            if 'video' in clean_msg: clean_msg = '<video>'
            if 'imagen' in clean_msg: clean_msg = '<imagen>'
            
        # Convert letters to lower case
        clean_msg = clean_msg.lower()
        
        # Iterate over characters in the message        
        for char in clean_msg:            
            # Remove special characters
            if not(char.islower() or char in [' ','_',':',"'",'<','>'] or char in emoji.UNICODE_EMOJI): 
                clean_msg = clean_msg.replace(char, "")
            
            # If emoji, add a space before the emoji
            if char in emoji.UNICODE_EMOJI:
                clean_msg  = emoji.demojize(clean_msg.replace(char,' ' + char))
        
        if "::" in clean_msg: clean_msg = clean_msg.replace('::',': :')
        
        print(clean_msg)
        # Create a list of words
        number_words.append(len(clean_msg.split()))
        clean_list.append(clean_msg.split()) 
```

```python
# Convert list of dictionaries to dataframe
df_msg = pd.DataFrame(list_messages)

# Filter rows of messages with content
content_true = [isinstance(elem, str) for elem in df_msg['Content']]
df_msg = df_msg[content_true].reset_index()

# Take only names and time
df_msg = df_msg[['Sender_name', 'TimeStamp']]

# Fix bad unicode characters in sender and change time format
df_msg['Sender_name'] = df_msg['Sender_name'].apply(fix_text)

# Append columns with the number of words per message
df_msg['number_words'] = number_words

# Display results
df_msg.head(10)
```

```python
# Flatten list of words
all_words = list(itertools.chain(*clean_list))

# Get list of stop words in Spanish
stop_words_sp = get_stop_words('spanish')
```

```python
# Repeat row as many times as words in the the message
df_msg_word = df_msg.set_index(['Sender_name'])['TimeStamp'].repeat(df_msg['number_words']).reset_index()

# Append column with flattened list of words
df_msg_word['word']  = all_words

# Append column indicator of emoji and stop word
df_msg_word['emoji'] = [':' in word for word in df_msg_word['word']]
df_msg_word['stop']  = [word in stop_words_sp for word in all_words]

# Display results
df_msg_word.head(20)
```

```python
# Save to csv
df_msg_word.to_csv('whatsapp_chat.csv', sep = '\t')
```
